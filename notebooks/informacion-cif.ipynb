# Extracción de la información disponible a partir de un CIF
Este cuaderno recoge la información relativa a los códigos de identificación fiscal (CIF) que identifican a las entidades jurídicas en España. Su creación viene regulada por en la [Orden de 3 de julio de 1998 por la que se modifica el anexo del Decreto 2423/1975, de 25 de septiembre, que regula el código de identificación de las personas jurídicas y entidades en general.](https://www.boe.es/eli/es/o/1998/07/03/(1))

Según este documento, el CIF incluye información relativa a:
- Tipo de entidad: codificada en su primera letra
- Provincia: codificada en sus dos primeros dígitos tras la letra
- Paises y territorios: que se identifican entidades de otros paises.
  
Esta información se modifico a posteriori en la [ORDEN EHA/451/2008, de 20 de febrero, por la que se regula la composición del número de identificación fiscal de las personas jurídicas y entidades sin personalidad jurídica.](https://www.boe.es/boe/dias/2008/02/26/pdfs/A11374-11376.pdf)

Dicha orden entró en vigor a día 1 de julio de 2008. En este documento vamos a recoger los cambios recogidos en esta orden y las implicaciones respecto al tratamiento de los datos.

## Clave de formas jurídicas y clases de entidades:
El primer dígito es el que nos ofrece información sobre el tipo de entidad, según la orden:

> El primer dígito se destinaba a indicar la forma jurídica de la sociedad o clase de entidad, recogiendo en un anexo una serie de claves de forma jurídica o clase de entidad que eran igualmente aplicables a entidades españolas o extranjeras.

Según el artículo 3. de la EHA/451/2008, _Claves sobre la forma jurídica de entidades españolas_ podemos encontrar las siguientes asignaciones según su forma jurídica:

A. Sociedades anónimas.
B. Sociedades de responsabilidad limitada.
C. Sociedades colectivas.
D. Sociedades comanditarias
E. Comunidades de bienes y herencias yacentes.
F. Sociedades cooperativas.
G. Asociaciones.
H. Comunidades de propietarios en régimen de pro-
piedad horizontal.
J. Sociedades civiles, con o sin personalidad jurídica.
P. Corporaciones Locales.
Q. Organismos públicos.
R. Congregaciones e instituciones religiosas.
S. Órganos de la Administración del Estado y de las
Comunidades Autónomas.
U. Uniones Temporales de Empresas.
V. Otros tipos no definidos en el resto de claves.

## Claves de provincias
Respecto a las claves de provincia encontramos la siguiente información:

> El código de identificación de las entidades españolas se completaba con dos dígitos destinados a contener un indicador de la provincia de su domicilio social en el momento de su constitución, cinco dígitos destinados a contener un número secuencial dentro de cada provincia y un dígito de control.

Por lo tanto, los dígitos en las posiciones dos y tres, nos informanrían respecto a la provincia en la que se constituyó la entidad. Sin embargo el formato de generación de los CIF varió con esta orden y acorde a las disposición general, a partir del 1 de Julio de 2008:

> Por último, en la configuración del número de identificación fiscal desaparece, en el caso de entidades españolas, la información relativa a la provincia de su domicilio social en el momento de su constitución y, en el caso de entidades extranjeras, la información relativa al territorio de su domicilio social.

En la actualidad (05/04/2022), los códigos CIF se generan siguiendo el Artículo 2. _Composición del número de identificación fiscal de las personas jurídicas y entidades sin personalidad jurídica._

> El número de identificación fiscal de las personas jurídicas y entidades sin personalidad jurídica estará compuesto por nueve caracteres, con la siguiente composición:
  a) Una letra, que informará sobre la forma jurídica, si se trata de una entidad española, o, en su caso, el carácter de entidad extranjera o de establecimiento permanente de una entidad no residente en España.
  b) Un número aleatorio de siete dígitos.
  c) Un carácter de control.

Por lo tanto, no es posible determinar la provincia de un entidad creada con posterioridad a 1 de Julio de 2018 a partir de sus dígitos dos y tres.

## Clave de paises y territorios
Según el artículo 4. de EHA/451/2008, _Clave de entidad extranjera_
> Para las personas jurídicas y entidad sin personalidad jurídica que carezcan de la nacionalidad espaola, el número de identificación fiscal comenzará con la letra N, que indicará su carácter de entidad extranjera.

La orden incluye también información relativa a una nueva clave:
> La clave N, que hasta ahora se asignaba a las entidades no residentes, se utilizará para informar sobre el carácter de entidad extranjera y se define una nueva clave, la W, para informar sobre el caracter de establecimiento permanente de una entidad no residente.

# Cambios introducidos en la orden EHA/451/2008
Según la orden, encontramos información contextual respecto al cambio de claves sobre la forma jurídica de entidades españolas.

La clave G, que correspondía a "asociaciones y otro tipo no definido" se desdobla en cuatro claves:
- G para asociaciones
- U para las uniones temporales de empresas
- J para sociedades civiles
- V para tipos no definidos en el resto de claves
  
La clave Q también se desdobla quedando:
- Q para organismos públicos
- R para Congregaciones e instituciones religiosas.

En la disposicion transitoria única encontramos la siguiente información:

> Durante el año 2008 la Agencia Estatal de Administración Tributaria comunicará a las personas iurídicas y entidades sin personalidad jurídica que, de acuerdo con lo dispuesto en esta orden, deban tener una letra distinta, el nuevo número de identificación fiscal que les corresponda.
No obstante, las entidades que entiendan que, conforme a lo dispuesto en esta orden ministerial, les corresponde otra distinta, podrán solicitar a la Agencia Estatal de Administración Tributaria, a partir del 1 de julio de 2008, la asignación de un nuevo número de identificación fiscal, sin esperar a recibir la comunicación indicada en el párrafo anterior.
Las entidades que a 1 de enero de 2009 no hayan recibido la comunicación indicada en el primer párrafo de esta disposición y, conforme a su forma jurídica, consideren que les debería corresponder un número de identificación fiscal distinto del que tienen asignado, deberán dirigirse a la Administración o Delegación de la Agencia Estatal de Administración Tributaria correspondiente a su domicilio fiscal para actualizar la información censal que tiene la Administración Tributaria de la entidad y obtener el número de identificación fiscal que, en su caso, les corresponda.

## Nota importante respecto al tratamiento automatizado
Según el artículo 1. de EHA/451/2008, _Invariabilidad del número de identificación discal de las personas jurídicas y entidades sin personalidad jurídica:

> La Administración tributaria asignará a las personas jurídicas y entidades sin personalidad jurídica un número de identificación fiscal que las identifique, y que **será invariable cualesquiera que sean las modificaciones que experimenten aquellas, salvo que cambie su forma jurídica o nacionalidad.**

¿Cómo afectan por tanto estos cambios al tratamiento de los datos?

### Clave de formas jurídicas
Las entidades de tipo unión temporales de empresas, sociedades civiles, o aquellas cuyo tipo no esté definido el resto de claves habrán cambiado su CIF a lo largo de 2008, reemplazando su letra inicial G por la letra U, J o V.

Las entidades de tipos congregación e institución religiosa, habrán cambiado su primera letra de Q a R a lo largo de 2008.

Aquellas entidades extranjeras que hayan pasado a ser residentes, habrán cambiado su letra de N a W a lo largo de 2008.

### Clave de provincia
En el caso de entidades que se crearon con anterioridad a 1 de Julio de 2008, los dígitos en las posiciones 2 y 3, podrían indicarnos la provincia en la que se constituyo la entidad.  A partir de de esta fecha, el CIF incluye siete dígitos aleatorios, que no están relacionados con la provincia.
